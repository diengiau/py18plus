# 1. What is `execucomp`?

This is a database about the US firms' executives and their compensation and benefits. The database starts since 1992 and covers S&P 1,500 firms. So each year, we have around 1,500 large firms.

As an example of the database, I will use year 2018 of Execucomp. Due to license, I don't provide the data for public. But you may go to the database room/library to check its availability.

The data will be stored as `csv` file and will be imported and analyzed using `pandas` library in Python.


```python
import pandas as pd
```

To import a csv file by `pandas`, use `read_csv(path_to_csv_file)` function:

```python
Execucomp = pd.read_csv("data/execucomp2018.csv")
```

First few lines/rows of data:

```python
Execucomp.head()
```

# 2. Discover the data

## 2.1 Filter

Usage: to check a subset of data.

For each firm, it will cover not only the CEO but also other directors. For example, in the database, __Microsoft Corp.__ (ticker `MSFT`) has 5 executives in 2018:

- Satya Nadella: CEO
- Amy E. Hood: CFO
- Bradford L. Smith: President and Chief Legal Officer
- Margaret L. Johnson: Executive Vice President of Business Development
- Jean-Philippe Courtois: Executive VP and President of Microsoft Global Sales, Marketing & Operations


```python
MSFT = Execucomp[Execucomp['ticker'] == "MSFT"]
MSFT
```

How about another high-tech firm: Apple Inc.?

```python
AAPL = Execucomp[Execucomp['ticker'] == "AAPL"]
AAPL
```

So in Apple board, we have a CEO (Tim), a CFO (Luca Maestri), a COO, one retail manager, and a general one.

## 2.2 Select

Usage: to select some variables only in the data.

Execucomp also offers the gender of each executive, let's see:

```python
Execucomp[Execucomp['ticker'].isin(["MSFT","AAPL"])][['ticker', 'exec_fullname', 'titleann', 'gender']]
```

Next, we care about the compensation of each executive. In Execucomp, we have a variable namely `tdc1`, which is the total compenstation in a year.

> Total compensation for the individual year, comprised of the following: Salary, Bonus, Other Annual, Total Value of Restricted Stock Granted, Total Value of Stock Options Granted (using Black-Scholes), Long-Term Incentive Payouts, and All Other Total.

```python
Execucomp[Execucomp['ticker'].isin(["MSFT","AAPL"])][['ticker', 'exec_fullname', 'titleann', 'gender', 'tdc1']]
```

Do you find something? Tim (CEO of Apple) receives less than other directors, but Satya Nadella	(CEO of MSFT) receives the highest compensation. Let's prove it by code.

## 2.3 Sort values

Usage: to sort a column by values.

```python
MSFT[['ticker', 'exec_fullname', 'titleann', 'gender', 'tdc1']].sort_values('tdc1')
```

```python
AAPL[['ticker', 'exec_fullname', 'titleann', 'gender', 'tdc1']].sort_values('tdc1')
```

## 2.4 create new variable

Now, we go further to see the composition of the total compensation (in thousand dollars).

```python
TwoFirm = Execucomp[Execucomp['ticker'].isin(["MSFT","AAPL"])][
    ['ticker', 'exec_fullname', 'titleann', 'gender',
    'tdc1', 'total_sec', 'salary', 'bonus', 'total_curr', 'stock_awards', "option_awards",
    "othcomp", 
    #'rstkgrnt', "option_awards_blk_value", "option_awards_rpt_value", 'allothtot', 'eip_unearn_val'
    #'opt_unex_exer_est_val', 'opt_unex_unexer_est_val', 
    ]
]
TwoFirm
```

How many percent the total compensation comes from (fixed) salary and equity parts?

```python
TwoFirm['salary_pct'] = TwoFirm['salary']/TwoFirm['tdc1']
TwoFirm['stock_pct'] = TwoFirm['stock_awards']/TwoFirm['tdc1']
TwoFirm
```

As you can see, most of the executive compensation is from equity (incentive) component, rather than the fixed salary. So, be careful whenever you read that a CEO received only $1 in a year.

## 2.5 `groupby` and summarize data

Sometimes, we wanna group data in groups:

- for each firm, find the highest amount of compensation
- for each firm, find the average salary
- for each firm, find the frequency of gender

Then, we may need the `groupby` function.

```python
TwoFirm.groupby('ticker')['tdc1'].agg(max)
```

```python
import numpy as np
TwoFirm.groupby('ticker')['tdc1'].agg(np.mean)
```

```python
TwoFirm.groupby('ticker')['gender'].value_counts()
```

# 3. Time series of compensation components

In this section, we will see how the total compensation components change over time across firms.



```python
CompComponent = pd.read_csv("data/CompComponent.csv")
```

```python
CompComponent
```

We draw a plot of these time series:

```python
import matplotlib.pyplot as plt

plt.plot('year', 'salary_pct', data=CompComponent, marker='', color='black', linewidth=2,label="salary")
plt.plot('year', 'stock_pct', data=CompComponent, marker='x', color='blue', linewidth=2, label="stock")
plt.plot('year', 'option_pct', data=CompComponent, marker='o', color='olive', linewidth=2, linestyle='dashed', label="option")
plt.legend()
```
